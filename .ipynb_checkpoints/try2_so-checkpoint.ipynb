```python
from numpy import *
from random import shuffle
def labelTransfor(labels):
    oneVec = labels.ravel()
    typeSet = set(oneVec.tolist())
    types = len(typeSet) #一共有多少个种类
    
    labels_new = zeros([labels.shape[0],types])
    for i in arange(labels.shape[0]):
        #print(labels[i][0])
        num = labels[i][0] #种类单个数字表示 如，8
        labels_new[i][num] = 1
    return labels_new
        
class DataObj(object):
    def __init__(self,trainImages,testImages,trainLabels,testLabels):
        self.trainImages=trainImages
        self.testImages=testImages
        self.trainLabels=trainLabels
        self.testLabels=testLabels
        
        self.var = 1
        self.maxVar = 1
        self.epoch=0
        num = self.trainImages.shape[0]
        self.num_index = arange(num)
        shuffle(self.num_index)
    def nextBatch(self,batch_size):
        self.maxVar = int(trainImages.shape[0]/batch_size)-1
        if self.var <= self.maxVar:
            return self.getBatch(batch_size)
        else :
            shuffle(self.num_index)
            self.epoch+=1
            self.var=1
            return self.getBatch(batch_size)
    def getBatch(self,batch_size):
        images_batch = zeros((batch_size,trainImages[0].shape[0]),float32)
        labels_batch = zeros((batch_size,trainLabels[0].shape[0]),float32)
        tup=(images_batch,labels_batch)
        for i in arange(batch_size):
            index = self.num_index[i + self.var*batch_size]
            #print("index:",index)
            tup[0][i] = trainImages[index]#index 是60091
            tup[1][i] = trainLabels[index]
        print("self.var",self.var)
        self.var+=1
        return tup
    
def getImagesLabels(lujing,sum_num,num):    #sum_num 一共多少个视频文件夹 num 每个视频提多少个特征
    classlist =  os.listdir(lujing)    #将basepath下的文件遍历并存入classlist
    
    #num=5  #每个视频特征文件夹中取多少个特征 c3d文件夹下一共sum_num个视频文件夹
    images=numpy.empty([sum_num,num,4096,1],numpy.uint8)
    labels=numpy.zeros([sum_num,10])
    
    for classvar in classlist:  #for output/ 下的文件夹 每个文件夹代表一个视频 sum_num次循环
       
        #获取完整路径 以及poi_1 第多少个视频文件夹/sum_num
        videodir = os.path.join('%s%s' % (lujing,classvar))#sum_num个
        poi_1=classlist.index(classvar)
        print("第 ",poi_1," 个视频文件夹")
        print(videodir)
        if poi_1==sum_num:
            break
        
        if os.path.isdir(videodir) == True:    #获取labels
            mid_2=numpy.zeros((10),numpy.int8)
            mode = re.compile(r'\d+')
            str = mode.findall(classvar)
            if(str):
                id_temp=str[0]
                if id_temp[0]=='0':#
                    id_temp=int(id_temp[1])
                #print(id_temp)
                mid_2[id_temp-1]=1
                print("labels：",mid_2)
            else:
                print("ERROR label exception")
            
            
            featlist=os.listdir(videodir)
            featlist.sort();
            mid_1=numpy.zeros((num,4096,1))
            
            for feat in featlist[:num] :#for 每一个视频文件夹中所有的.fc6feadirlist 
                poi_2=featlist.index(feat)
                print("   第 ",poi_2," 个特征文件     判断是否.fc6-1为结尾")
                pattern='.+\.(fc6-1)$'
                m=re.match(pattern,feat)
                if m:#符合
                    print("      符合 ，为",feat)
                    temp_perFeatDir = os.path.join('%s%s/%s' % (lujing,classvar,feat))#每个fc6路径
                    a=numpy.fromfile(temp_perFeatDir,dtype=numpy.uint8)#读取fc6
                    numpy.savetxt("temp.txt",a[5:4101].reshape(4096,1),fmt="%d")
                    mid_1[poi_2]=a[5:4101].reshape(4096,1)
                    numpy.savetxt("temp.txt",a[5:4101].reshape((1,4096)),fmt="%d")
                    with open('temp.txt','r') as f1:
                        with open('trains.txt','w+') as f2:
                            f2.write(f1.read())
                else:#不符合
                    print("     不符合 ，为",feat)
                #print("$$$$$$$$mid_1",mid_1)
                #新建文件夹目录 新建每个视频的文件夹
            images[poi_1]=mid_1
            labels[poi_1]=mid_2
        else:
            print("haha")
    #images.tofile("images.bin")
    return images,labels
```

```python
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets("MNIST_data/",one_hot = True)

print(mnist.train.images.shape, mnist.train.labels.shape)

print(mnist.test.images.shape,mnist.test.labels.shape)

print(mnist.validation.images.shape,mnist.validation.labels.shape)

import tensorflow as tf

sess = tf.InteractiveSession()
x = tf.placeholder(tf.float32,[None,784])

W = tf.Variable(tf.zeros([784,10]))
b = tf.Variable(tf.zeros([10]))

y = tf.nn.softmax(tf.matmul(x,W)+b) 

y_=tf.placeholder(tf.float32,[None,10])

cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_*tf.log(y),reduction_indices=[1]))

train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)

tf.global_variables_initializer().run()

for i in range(1000):
    batch_xs,batch_ys = mnist.train.next_batch(100)
    train_step.run({x: batch_xs,y_:batch_ys})

correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(y_,1))

accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))

print(accuracy.eval({x:mnist.test.images,y_:mnist.test.labels}))

```
